```python
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder


# Dictionary to store the accepted values for each categorical column
accepted_values = {
    'cap-shape': ['x', 'f', 's', 'b', 'o', 'p', 'c'],
    'cap-surface': ['t', 's', 'y', 'h', 'g', 'd', 'k', 'e', 'i', 'w', 'l'],
    'cap-color': ['n', 'y', 'w', 'g', 'e', 'o', 'p', 'r', 'u', 'b', 'k', 'l'],
    'does-bruise-or-bleed': ['f', 't'],
    'gill-attachment': ['a', 'd', 'x', 'e', 's', 'p', 'f'],
    'gill-spacing': ['c', 'd', 'f'],
    'gill-color': ['w', 'n', 'y', 'p', 'g', 'o', 'k', 'f', 'r', 'e', 'b', 'u'],
    'stem-root': ['b', 's', 'r', 'c'],
    'stem-surface': ['s', 'y', 'i', 't', 'g', 'k', 'h', 'f'],
    'stem-color': ['w', 'n', 'y', 'g', 'o', 'e', 'u', 'p', 'k', 'r', 'l', 'b', 'f'],
    'veil-color': ['w', 'y', 'n', 'u', 'k', 'e'],
    'has-ring': ['f', 't'],
    'ring-type': ['f', 'e', 'z', 'l', 'r', 'p', 'g', 'm'],
    'spore-print-color': ['k', 'p', 'w', 'n', 'r', 'u', 'g'],
    'habitat': ['d', 'g', 'l', 'm', 'h', 'w', 'p', 'u', 'e', 's'],
}

# Function to replace invalid categorical values
def replace_invalid_values(df, column, accepted_values):
    return df[column].apply(lambda x: x if x in accepted_values else f'created_{column}')

# Apply the function to categorical columns
for column, values in accepted_values.items():
    df[column] = replace_invalid_values(df, column, values)

for column, values in accepted_values.items():
    test[column] = replace_invalid_values(test, column, values)

```

```python
simple_imputer = SimpleImputer(strategy='mean')
miss_num = ['cap-diameter']

# Define the OneHotEncoder for categorical columns
onehot_encoder = OneHotEncoder(drop = 'first')

# Create a ColumnTransformer to apply SimpleImputer and OneHotEncoder
preprocessor = ColumnTransformer(
    transformers=[
        ('imputer', simple_imputer, miss_num),  # Apply imputer to categorical columns
        ('onehot', onehot_encoder, categorical_columns)    # Apply one-hot encoding to categorical columns
    ],
    remainder='passthrough'  # Leave the rest of the columns unchanged
)

# Create a pipeline that first imputes missing values, then applies one-hot encoding
pipeline = Pipeline(steps=[('preprocessor', preprocessor)])

# Fit and transform the data
X_processed = pipeline.fit_transform(X)

# Convert the result back to a DataFrame if needed
#X_processed = pd.DataFrame(X_processed, columns=pipeline.named_steps['preprocessor'].get_feature_names_out())
```
